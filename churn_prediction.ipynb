```python
from sklearn.model_selection import train_test_split
from tensorflow  import keras
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\rudra\\Downloads\\WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.drop(["customerID"],axis="columns",inplace=True)
```

```python
col=[]
col.append(df.columns)
```

```python
result=[]
for i in col:
    result.append(df[i].isnull().sum())

result
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
df["gender"]=le.fit_transform(df["gender"])
df["Partner"]=le.fit_transform(df["Partner"])
df["Dependents"]=le.fit_transform(df["Dependents"])
df["PhoneService"]=le.fit_transform(df["PhoneService"])
```

```python
df["OnlineSecurity"]=le.fit_transform(df["OnlineSecurity"])
df["OnlineBackup"]=le.fit_transform(df["OnlineBackup"])
df["DeviceProtection"]=le.fit_transform(df["DeviceProtection"])
df["TechSupport"]=le.fit_transform(df["TechSupport"])
df["StreamingTV"]=le.fit_transform(df["StreamingTV"])
df["StreamingMovies"]=le.fit_transform(df["StreamingMovies"])
df["PaperlessBilling"]=le.fit_transform(df["PaperlessBilling"])
df["Churn"]=le.fit_transform(df["Churn"])
df
```

```python
dummies_MultipleLines=pd.get_dummies(df["MultipleLines"]).astype(int)
dummies_InternetService=pd.get_dummies(df["InternetService"]).astype(int)
dummies_Contract=pd.get_dummies(df["Contract"]).astype(int)
dummies_PaymentMethod=pd.get_dummies(df["PaymentMethod"]).astype(int)
df=pd.concat([df,dummies_Contract,dummies_MultipleLines,dummies_InternetService,dummies_PaymentMethod],axis="columns")
df=df.drop(["MultipleLines","InternetService","Contract","PaymentMethod"],axis='columns')
df
```

```python
df.shape
```

```python
# df["tenure"]=df["tenure"]/100
```

```python
X=df.drop(['Churn'],axis='columns')
y=df.Churn
```

```python
X["TotalCharges"] = pd.to_numeric(X["TotalCharges"] , errors='coerce')
X=np.asarray(X).astype(int)
X
```

```python
# X["TotalCharges"].isin([' '])
# X.dtypes
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,)),
    keras.layers.Dense(300,activation="relu"),
    keras.layers.Dense(100,activation="swish"),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(loss='binary_crossentropy',optimizer="adam",metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=50)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
```
