```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("C:\\Users\\rudra\\Pictures\\IMG_20230926_110136.jpg")
```

```python
edges=cv2.Canny(img,95,90,L2gradient=False)
plt.imshow(edges)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
plt.imshow(th2)
```
